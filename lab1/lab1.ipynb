```python
import random as random
import statistics as s
import matplotlib.pyplot as pt
a = []
for i in range(10):
    a.append(random.random())
print(a)
med = s.median(a)
sr = sum(a)/len(a)
print(med, sr)

pt.grid()
pt.scatter(a, [i for i in range(1, 11)])
```

```python
import math as m
import matplotlib.pyplot as pt
a = []
for x in range(1, 11):
    y = (m.sqrt(1+(m.e)**(m.sqrt(x))) + m.cos(x**2))/(abs(1-m.sin(x)**3)) + m.log(2*x)
    a.append(y)
b = a[:5]
print(b)
pt.plot(a)
pt.show()
g = [i for i in range(1, 11)]
pt.scatter([i for i in range(1, 6)], b)
pt.show()
```
