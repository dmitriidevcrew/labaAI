```python
# EXAMPLE 1.2.3

import numpy as np
def euclidean_distance(v1, v2):
    return sum((x - y) ** 2 for x, y in zip(v1, v2)) ** 0.5
x = np.array([0, 0, 0])
y = np.array([3, 3, 3])
print(euclidean_distance(x, y))
```

```python
# EXAMPLE 1.2.4

def sqr_euclidean_distance(v1, v2):
    return sum((x - y) ** 2 for x, y in zip(v1, v2))

def weighted_euclidean_distance(v1, v2, w):
    return sum((x - y) ** 2 * s for x, y, s in zip(v1, v2, w)) ** 0.5

def manhattan_distance(v1, v2):
    return sum(abs(x-y) for (x, y) in zip(v1, v2))

def chebyshev_distance(v1, v2):
    return max(abs(x-y) for (x, y) in zip(v1, v2))

x = np.array([0, 0, 0])
y = np.array([3, 3, 3])
w = np.array([0, 0, 1])

print(sqr_euclidean_distance(x, y))
print(weighted_euclidean_distance(x, y, w))
print(manhattan_distance(x, y))
print(chebyshev_distance(x, y))
```

```python
# EXERCISE 1.3.1
x = np.array([0, 0, 0])
y = np.array([1, 1, 1])
z = np.array([2, 3, 4])
w = np.array([1, 5, 5])
a = [x, y, z, w]
def evklid_distance(x, y):
    return np.linalg.norm(x-y)
def evklid_square(x, y):
    return np.linalg.norm(x-y)**2
def chebyshev(x, y):
    return np.linalg.norm(x - y, ord=np.inf)
def hemming(x, y):
    return np.linalg.norm(x - y, ord=1)
def output(a1, a2):
    print(": {a1}, {a2}".format(a1=a1, a2=a2))
    print("  = {d1}\n  = {d2}\n  = {d3}\n  = {d4}\n".format(d1=evklid_distance(a1, a2), d2=evklid_square(a1, a2), d3=chebyshev(a1, a2), d4=hemming(a1, a2)))

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

for i in range(0, len(a)):
    ax.scatter(a[i][0], a[i][1], a[i][2])
    for j in range(i+1, len(a)):
        output(a[i], a[j])
plt.show()
```

```python
# EXAMPLE 2.2.1
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# 
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
target = [0, 0, 0, 1, 1, 1]

#   k-   
K = 3
model = KNeighborsClassifier(n_neighbors = K)
model.fit(X, target)
print(model)

#  
print( '(-2,-2) is class'),
print( model. predict([[-2,-2]]) )

print( '(1,3) is class'),
print( model.predict([[1,3]]) )
```

```python
# EXERCISE 2.3.1
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

iris = sns.load_dataset('iris')
iris

def init_model(k, X_train, y_train, X_test): #      (k = 1, 5, 10).
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return y_pred

def graph(y_test, X_test, y_pred): #   
    plt.figure(figsize=(10, 7))
    sns.scatterplot(x='petal_width', y='petal_length', data=iris, hue='species', s=70)
    plt.xlabel(" , ")
    plt.ylabel(" , ")
    plt.legend(loc=2)
    plt.grid()

    for i in range(len(y_test)):
        if np.array(y_test)[i] != y_pred[i]:
            plt.scatter(X_test.iloc[i, 3], X_test.iloc[i, 2], color='red', s=150)

X_train, X_test, y_train, y_test = train_test_split(
    iris.iloc[:, :-1],
    iris.iloc[:, -1],
    test_size = 0.15 #15%
)

X_train.shape, X_test.shape, y_train.shape, y_test.shape

X_train.head()
y_train.head()

k = int(input('  k'))

y_pred = init_model(k, X_train, y_train, X_test)
print((f'accuracy: {accuracy_score(y_pred, y_test) :.3}'))
graph(y_test, X_test, y_pred)
```
